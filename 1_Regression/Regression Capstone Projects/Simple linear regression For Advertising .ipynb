**MediaMax Advertising faces a critical challenge in optimizing their advertising budget across TV, Radio, and Newspaper channels to maximize sales.**

### Challenges:
1. **Limited Budget**: $100,000 monthly advertising spend
2. **3 Main Channels**: Television (TV), Radio, Newspaper
3. **Unclear ROI**: Some channels deliver strong results, others waste money
4. **Need Data-Driven Decision**: Where to allocate budget for maximum return?

### Available Data:
```
📊 200 monthly sales observations
📈 Predictors: TV, Radio, Newspaper spending (in $000s)
🎯 Target: Sales (in $000s units)
```

### Key Research Question:
**"How should the advertising budget be allocated across TV, Radio, and Newspaper to achieve maximum sales?"**

### Objectives:
- Build accurate predictive model for sales
- Identify most impactful advertising channels

````python
def gradient_descent(x, y, iterations):
    x = x.values if hasattr(x, 'values') else np.array(x)
    y = y.values if hasattr(y, 'values') else np.array(y)
    m = 0; c = 0; L = 0.001  # Learning rate
    
    for i in range(iterations):
        # COST FUNCTION
        y_pred = m * x + c
        cost = (1/len(x)) * sum((y - y_pred)**2)
        
        # GRADIENT DESCENT UPDATES (MISSING!)
        dm = (2/len(x)) * sum(x * (y_pred - y))  # Slope derivative
        dc = (2/len(x)) * sum(y_pred - y)        # Intercept derivative
        
        m = m - L * dm  # Update slope
        c = c - L * dc  # Update intercept
        
        # Print every 100 iterations
        if i % 100 == 0:
            print(f"Iteration {i} | Cost = {cost:.4f} | m = {m:.4f} | c = {c:.4f}")

    plt.scatter(x, y, color='blue', label='Data')
    plt.plot(x, m * x + c, color='red', label=f'y = {m:.2f}x + {c:.2f}')
    plt.xlabel("x"); plt.ylabel("y")
    plt.title("Linear Regression using Gradient Descent")
    plt.legend(); plt.show()
    
    return m, c
````
