```python
# Set style for better plots
plt.style.use('default')
sns.set_palette("husl")

print("📊 Advertising Dataset - Complete Matplotlib Profiling")
print(f"Dataset shape: {df.shape}")

# 1. CORRELATION HEATMAP (Most Important)
plt.figure(figsize=(10, 8))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, 
            square=True, linewidths=0.5)
plt.title('Advertising Spend vs Sales - Correlation Matrix', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()

print("\n🔥 Key Insights:")
print("TV-Sales correlation: {:.2f} (STRONGEST)".format(correlation_matrix.loc['Sales', 'TV']))
print("Radio-Sales: {:.2f}".format(correlation_matrix.loc['Sales', 'Radio']))
print("Newspaper-Sales: {:.2f}".format(correlation_matrix.loc['Sales', 'Newspaper']))

# 2. COMPREHENSIVE SUBPLOTS
fig, axes = plt.subplots(2, 3, figsize=(18, 12))
fig.suptitle('Complete Advertising Dataset Analysis', fontsize=16, fontweight='bold')

# Distribution plots
axes[0,0].hist(df['TV'], bins=20, color='red', alpha=0.7, edgecolor='black')
axes[0,0].set_title('TV Spend Distribution')
axes[0,0].set_xlabel('TV Spend ($K)')

axes[0,1].hist(df['Radio'], bins=20, color='green', alpha=0.7, edgecolor='black')
axes[0,1].set_title('Radio Spend Distribution')
axes[0,1].set_xlabel('Radio Spend ($K)')

axes[0,2].hist(df['Newspaper'], bins=20, color='blue', alpha=0.7, edgecolor='black')
axes[0,2].set_title('Newspaper Spend Distribution')
axes[0,2].set_xlabel('Newspaper Spend ($K)')

# Scatter plots vs Sales
axes[1,0].scatter(df['TV'], df['Sales'], alpha=0.6, color='red')
axes[1,0].set_title('TV vs Sales (r={:.2f})'.format(correlation_matrix.loc['Sales', 'TV']))
axes[1,0].set_xlabel('TV Spend')
axes[1,0].set_ylabel('Sales')

axes[1,1].scatter(df['Radio'], df['Sales'], alpha=0.6, color='green')
axes[1,1].set_title('Radio vs Sales (r={:.2f})'.format(correlation_matrix.loc['Sales', 'Radio']))
axes[1,1].set_xlabel('Radio Spend')
axes[1,1].set_ylabel('Sales')

axes[1,2].scatter(df['Newspaper'], df['Sales'], alpha=0.6, color='blue')
axes[1,2].set_title('Newspaper vs Sales (r={:.2f})'.format(correlation_matrix.loc['Sales', 'Newspaper']))
axes[1,2].set_xlabel('Newspaper Spend')
axes[1,2].set_ylabel('Sales')

plt.tight_layout()
plt.show()

# 3. BOX PLOTS (Outliers detection)
plt.figure(figsize=(12, 6))
df.boxplot(column=['TV', 'Radio', 'Newspaper', 'Sales'])
plt.title('Box Plots - Outlier Detection')
plt.ylabel('Values ($K)')
plt.xticks(rotation=45)
plt.show()

print("\n✅ Complete Matplotlib Profiling Complete!")
print("📈 Ready for Linear Regression modeling!")
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
# Your current data
X = df[['TV', 'Radio']]  # (200, 2)
y = df['Sales']          # (200,)

print("X shape:", X.shape)    # (200, 2)
print("y shape:", y.shape)    # (200,)

# Reshape y to 2D
y_2d = y.values.reshape(-1, 1)  # (200, 1)
print("y_2d shape:", y_2d.shape)

# Now both X and y_2d are 2D ✓
lm_model = LinearRegression()
lm_model.fit(X, y_2d)
```

```python
lm_model.coef_
```

```python
lm_model.intercept_
```

```python
lm_model = LinearRegression()
lm_model.fit(X, y)
lm_model.score(X, y)
```

```python
print("Intercept:", lm_model.intercept_)
print("Slope:", lm_model.coef_[0])
r2_score = lm_model.score(X, y)
print(f"R² Score: {r2_score:.3f}")
```
